```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df=pd.read_csv(r"C:\Users\Admin\Downloads\possum.csv")
df
```

```python
df["age"]=df["age"].fillna(df["age"].mean())
df["footlgth"]=df["footlgth"].fillna(df["footlgth"].mean())
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns=['totlngth','sex',"Pop"])
y = df['totlngth']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
r2_score(y_train, y_pred_train),  r2_score(y_test, y_pred_test),  
```

```python
mean_absolute_error(y_train, y_pred_train),  mean_absolute_error(y_test, y_pred_test),  
```

```python
mean_squared_error(y_train, y_pred_train),  mean_squared_error(y_test, y_pred_test),  
```
